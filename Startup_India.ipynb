```python
#importing the required library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset
myData=pd.read_csv('Startup_India.csv')
myData
```

```python
#assiging some dummy values to state data
dummyData=pd.get_dummies(myData.State)
dummyData
```

```python
#merging the dataset with dummydata
mergedummyData=pd.concat([myData,dummyData],axis='columns')
mergedummyData
```

```python
#final dataset values
finaldummyData=mergedummyData.drop(['State','Mumbai'],axis='columns')
finaldummyData
```

```python
#assiging the x values
x=finaldummyData.drop(['Profit'],axis='columns')
x
```

```python
#assiging the y values
y=finaldummyData.Profit
y
```

```python
from sklearn.linear_model import LinearRegression
createmodel=LinearRegression()
```

```python
#creating linear model
createmodel.fit(x,y)
```

```python
#predicting the values
createmodel.predict([[12349876,123456,45678,0,0]])
```

```python
#checking how perfect is our model
createmodel.score(x,y)
```
